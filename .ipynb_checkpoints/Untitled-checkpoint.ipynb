```python
df = pd.read_excel('corpus_indonesia.xlsx')
```

```python
text = list(df['Text'])
trait = list(df['Traits'])
```

```python
res = []
for i in range(len(text)):
    res.append((text[i], trait[i]))
```

```python
import re

sent = 'saya benci tapi rindu'
sent = sent.lower()
tokens = sent.split()

list_traits = ['Aggreeableness', 'Conscientiousness', 'Extraversion', 'Neuroticism', 'Openness']
list_freq = [0, 0, 0, 0, 0]
i = 0

while i < len(tokens):
    temp = []
    temp_traits = []
    cur.execute("SELECT * FROM ontologi WHERE words like '%" + tokens[i] + "%'")
    for row in cur.fetchall():
        temp.append(row[1])
        temp_traits.append(row[2])
    maks = 0
    trait = ''
    for k in range(len(temp)):
        if re.sub('_', ' ', temp[k].lower()) in sent:
            if len(temp[k].split('_')) > maks:
                trait = temp_traits[k]
                maks = len(temp[k].split('_'))
    if maks > 0:
        list_freq[list_traits.index(trait)] += 1
        i += maks
    else:
        i += 1
```
